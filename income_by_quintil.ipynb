```r
equity = read_excel("Data_Extract_From_Poverty_and_Equity.xlsx")   
```

```r
equity = data.frame(equity)
tail(equity)
```

```r
columns = colnames(equity)
columns = str_replace(columns, "\\.", "")
columns = str_replace(columns, "\\.", "")
columns = str_replace(columns, "\\.", "")
columns = str_sub(columns, start = -4)
columns[1] = "series_name"
columns[2] = "series_code"
columns[3] = "country_name"
columns[4] = "country_code"
columns
```

```r
colnames(equity) = columns
tail(equity)
```

```r
equity = equity %>%
    mutate_at(columns[5:length(columns)], as.numeric) %>%
    melt() %>%
    mutate(year = as.character(variable)) %>%
    na.omit
tail(equity)
```

```r
most_recent_info = equity %>%
#     mutate(year = as.character(variable)) %>%
    select(country_code, year) %>%
#     mutate_at(c('variable'), as.numeric) %>%
#     select(country_code, variable) %>%
    group_by(country_code) %>%
    summarize(year = max(year))
tail(most_recent_info)
```

```r
most_recent_info = equity %>%
    inner_join(most_recent_info, by = c("year", "country_code")) %>%
    group_by(country_code, year) %>%
    summarize(total = sum(value)) %>%
    inner_join(equity, by = c("year", "country_code")) %>%
    filter(series_name == "Income share held by highest 20%") %>%
    mutate(first_quintil = value / total) %>%
    distinct(country_code, year, total, first_quintil)

tail(most_recent_info)
```

```r
final_dataset = equity %>%
    inner_join(most_recent_info, by = c("year", "country_code")) %>%
    mutate(country_name = paste(country_name, " (", year, ")", sep="" )) %>%
    select(series_name, country_name, value, first_quintil, total) %>%
    mutate(percentage_of_total = value / total) %>%
    mutate(series_name = replace(series_name, series_name=="Income share held by highest 20%", "1 - Top 20%")) %>%
    mutate(series_name = replace(series_name, series_name=="Income share held by second 20%", "2 - 20%-40%")) %>%
    mutate(series_name = replace(series_name, series_name=="Income share held by third 20%", "3 - 40%-60%")) %>%
    mutate(series_name = replace(series_name, series_name=="Income share held by fourth 20%", "4 - 60%-80%")) %>%
    mutate(series_name = replace(series_name, series_name=="Income share held by lowest 20%", "5 - Bottom 20%")) 
#     spread(series_name, value) 

tail(final_dataset)
```

```r
options(repr.plot.width=16, repr.plot.height=50)
final_dataset %>%
    ggplot(aes(x=forcats::fct_reorder(country_name, first_quintil),y=desc(percentage_of_total), 
               fill=series_name)) + 
    geom_col() + 
    coord_flip() + 
    geom_text(aes(label=paste0(sprintf("%1.1f", percentage_of_total*100),"%")), 
             position=position_stack(vjust=0.5)) +
    labs(x="Country (most recent year info)", y="Share of Income", 
         caption = c("Data Source: https://databank.worldbank.org/source/poverty-and-equity", "Source code: ")) + 
        theme_minimal() +
    theme(
#         plot.caption = element_text(hjust=c(1, 0)),
        axis.text.x=element_blank(),
        axis.ticks.x=element_blank(),
    )
    
#     annotate("text",  x=-20, y = 0, label = "Data Source: https://databank.worldbank.org/source/poverty-and-equity", 
#              fontface = "bold") + 

```
